# Preprocessing 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#To load the dataset
df=pd.read_csv("C:/Users/INTERFACE/OneDrive/Desktop/ENTRI ASSIGNMENT/PYTHON/myexcel - myexcel.csv.csv")
df
```

```python
#To correct the data in the "height" column by replacing it with random numbers between 150 and 180.
df['Height'] = np.random.randint(150, 181, size=len(df))
df
```

```python
#To ensure data consistency and integrity 
df.drop_duplicates(inplace=True)
df.info()
df.head()
```

# Analysis Tasks

```python
#1.Distribution of employees across each team with percentage
team_distribution = df['Team'].value_counts().reset_index()
team_distribution.columns = ['Team', 'Employee Count']
team_distribution['Percentage'] = (team_distribution['Employee Count'] / len(df)) * 100
print("Employee Distribution by Team:\n", team_distribution)
```

```python
#2.Segregating employees by their positions
position_distribution = df['Position'].value_counts().reset_index()
position_distribution.columns = ['Position', 'Employee Count']
print("Employee Distribution by Position:\n", position_distribution)
```

```python
#3.Predominant age group
age= [18, 22, 25, 28, 31, 34, 40]
age_labels = ['18-22', '23-25', '26-28', '29-31', '32-34', '35-40']
df['Age Group'] = pd.cut(df['Age'], bins=age, labels=age_labels)
age_group_distribution = df['Age Group'].value_counts().sort_index()
print("Predominant Age Group:\n", age_group_distribution)
```

```python
#4.Team and Position with Highest Salary Expenditure
salary_df = df.dropna(subset=['Salary'])
salary_summary = salary_df.groupby(['Team', 'Position'])['Salary'].sum().reset_index()
highest_salary = salary_summary.sort_values(by='Salary', ascending=False).head(1)
print("Team and Position with Highest Salary Expenditure:\n", highest_salary)
```

```python
# 5.Correlation between Age and Salary
corr_df = df.dropna(subset=['Salary'])
correlation = corr_df['Age'].corr(corr_df['Salary'])
print(f"Correlation Between Age and Salary: {correlation:.2f}")

#For visual representation
plt.figure(figsize=(10, 6))
sns.regplot(x='Age', y='Salary', data=corr_df, scatter_kws={'alpha':0.6})
plt.title('Correlation Between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid()
plt.show()
```

# Graphical Representations

```python
#1.Distribution of employees across each team with percentage 
plt.figure(figsize=(10, 6))
sns.barplot(y='Team', x='Employee Count', data=team_distribution.sort_values('Employee Count', ascending=False), palette='viridis')
plt.title('1. Employee Distribution by Team')
plt.xlabel('Number of Employees')
plt.ylabel('Team')
plt.show()
```

```python
#2.Segregate employees by their positions
plt.figure(figsize=(10, 6))
plt.pie(position_distribution['Employee Count'], labels=position_distribution['Position'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Set2"))
plt.title('2. Employee Distribution by Position')
plt.axis('equal')
plt.show()
```

```python
#3.Predominant age group
plt.figure(figsize=(10, 6))
sns.barplot(x=age_group_distribution.index, y=age_group_distribution.values, palette='muted')
plt.title('3. Employee Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.show()
```

```python
# 4.Team and Position with Highest Salary Expenditure
plt.figure(figsize=(12, 6))
sns.barplot(x='Salary', y=highest_salary['Team'] + ' - ' + highest_salary['Position'], data=highest_salary, palette='rocket')
plt.title('4. Top 10 Salary Expenditures by Team & Position')
plt.xlabel('Total Salary')
plt.ylabel('Team - Position')
plt.show()
```

```python
#5.Correlation between Age and Salary
plt.figure(figsize=(10, 6))
sns.regplot(x='Age', y='Salary', data=corr_df, scatter_kws={'alpha':0.6}, line_kws={'color': 'red'})
plt.title('5. Correlation Between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```
